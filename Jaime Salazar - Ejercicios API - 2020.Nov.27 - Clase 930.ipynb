### Ej2
Del campo "Report-To", extrae el valor de la clave "url", que te devuelve el API

```python
#Usando Serie
R1Ser=pd.Series(response.headers)

Ej2Ser=R1Ser
Ej2Ser["Report-To"]=json.loads(Ej2Ser["Report-To"])
Ej2Ser["Report-To"]["endpoints"]=Ej2Ser["Report-To"]["endpoints"][0]
Ej2Ser=Ej2Ser["Report-To"]["endpoints"]["url"]

#Usando DF
Ej2Df=R1Data

Ej2Df=json.loads(Ej2Df.loc["Report-To",0])
Ej2Df["endpoints"]=Ej2Df["endpoints"][0]
Ej2Df=Ej2Df["endpoints"]["url"]

print("Usando Serie: ",Ej2Ser)
print()
print("Usando DF: ",Ej2Df)
```

### Ej9
Usando la misma url (api.exchangeratesapi.io), extrae las cotizaciones de las diferentes divisas frente al CAD entre las fechas 1/10/2019 y 10/10/2019

```python
# Respuesta

#Con Payload
# Respuesta
payload = {'start_at': '2019-10-01', 'end_at': "2019-10-10","base":"CAD"}
Ej91 = requests.get("https://api.exchangeratesapi.io/history",params=payload)

print("Con Payload:")
print(pd.DataFrame(pd.Series(Ej91.json())).loc["rates",0])
```

```python
#Directo
Ej92 = requests.get("https://api.exchangeratesapi.io/history?start_at=2019-10-01&end_at=2019-10-10&base=CAD")

print("Directo:")
print(pd.DataFrame(pd.Series(Ej92.json())).loc["rates",0])
print()
```
